```python
#Ahora vamos a retirar los datos atípicos de las variables en cuestión
df = df[(df['Kg malta'] >= 17370.25) & (df['Kg malta'] <= 18700)]
df = df[(df['Bultos Malta caramelo'] >= 6) & (df['Bultos Malta caramelo'] <= 26)]
df = df[(df['Color malta caramelo [EBC]'] >= 109) & (df['Color malta caramelo [EBC]'] <= 165)]
#df = df[(df['Color cocimiento'] >= 11.4) & (df['Color cocimiento'] <= 17)]
df = df[(df['Color ponderado silo'] >= 3.214) & (df['Color ponderado silo'] <= 4.8)]
```

```python
#Ahora voy a retirar las columnas de lote y silo   
df = df.drop(columns=['Lote malta caramelo', 'Silo'], axis=1)
```

```python
#Ahora debo dividir los datos
from sklearn.model_selection import train_test_split
#Aquí saco la columna que voy a predecir
X = df.drop(['Color cocimiento'], axis=1)
#Aquí digo cuál es la columna que voy a predecir
Y = df['Color cocimiento']
#Aquí hago el split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train.head()
```

```python

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(max_depth=4, random_state=42)
model.fit(X_train,Y_train)
```

```python
Y_pred= model.predict(X_test)
Y_pred
```

```python
#ahora debo evaluar el modelo
from sklearn.metrics import mean_squared_error, r2_score

print(f'MSE: {mean_squared_error(Y_test, Y_pred)}')
print(f'R2: {r2_score(Y_test, Y_pred)}')

from math import sqrt
print(f'RMSE: {sqrt(mean_squared_error(Y_test, Y_pred))}')
#Ahora sacaré las mismas metricas con los datos de entrenamiento
Y_train_pred = model.predict(X_train)
print(f'MSE train: {mean_squared_error(Y_train, Y_train_pred)}')
print(f'R2 train: {r2_score(Y_train, Y_train_pred)}') 
print(f'RMSE train: {sqrt(mean_squared_error(Y_train, Y_train_pred))}')
```
