```python
#Ahora vamos a retirar los datos atÃ­picos de las variables en cuestiÃ³n
df = df[(df['Kg malta'] >= 17370.25) & (df['Kg malta'] <= 18700)]
df = df[(df['Bultos Malta caramelo'] >= 12) & (df['Bultos Malta caramelo'] <= 26)]
df = df[(df['Color malta caramelo [EBC]'] >= 109) & (df['Color malta caramelo [EBC]'] <= 165)]
#df = df[(df['Color cocimiento'] >= 11.4) & (df['Color cocimiento'] <= 17)]
df = df[(df['Color ponderado silo'] >= 3.214) & (df['Color ponderado silo'] <= 4.8)]
```

```python
#Ahora voy a retirar las columnas de lote y silo   
df = df.drop(columns=['Lote malta caramelo', 'Silo'], axis=1)
```

```python
from utils import analisis_boxplot
analisis_boxplot(df,'Color cocimiento')
#Se cuenta con los datos sin atÃ­picos no deseados.
```

```python
#Ahora debo dividir los datos
from sklearn.model_selection import train_test_split
#AquÃ­ saco la columna que voy a predecir
X = df.drop(['Color cocimiento'], axis=1)
#AquÃ­ digo cuÃ¡l es la columna que voy a predecir
Y = df['Color cocimiento']
#AquÃ­ hago el split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)

X_train.head()
```

```python
from xgboost import XGBRegressor

# Rename columns to remove special characters
X_train_clean = X_train.rename(columns=lambda x: x.replace('[', '').replace(']', '').replace('<', '').replace('>', '').replace(' ', '_'))
X_test_clean = X_test.rename(columns=lambda x: x.replace('[', '').replace(']', '').replace('<', '').replace('>', '').replace(' ', '_'))

model = XGBRegressor(random_state=42)

model.fit(X_train_clean, Y_train)
```

```python
Y_pred = model.predict(X_test_clean)
Y_pred
```

```python
#ahora debo evaluar el modelo
from sklearn.metrics import mean_squared_error, r2_score

print(f'MSE: {mean_squared_error(Y_test, Y_pred)}')
print(f'R2: {r2_score(Y_test, Y_pred)}')

from math import sqrt
print(f'RMSE: {sqrt(mean_squared_error(Y_test, Y_pred))}')
```

```python
#HiperparametrizaciÃ³n
from sklearn.model_selection import GridSearchCV

params = {
	'max_depth': [3, 4, 5],
	'learning_rate': [0.01, 0.1, 0.2],
	'n_estimators': [20, 50, 100, 200, 300]
}

grid = GridSearchCV(model, params, scoring='neg_mean_squared_error', cv=5, n_jobs=2)
grid.fit(X_train_clean, Y_train)

print(grid.best_params_)
print(grid.best_estimator_)

best_model = grid.best_estimator_
mean_squared_error(Y_test,best_model.predict(X_test_clean))
```

```python
# Hacer las predicciones
Y_best_pred = best_model.predict(X_test_clean)
Y_best_pred
```

```python
#ahora debo evaluar el modelo
from sklearn.metrics import mean_squared_error, r2_score

print(f'MSE: {mean_squared_error(Y_test, Y_best_pred)}')
print(f'R2: {r2_score(Y_test, Y_best_pred)}')

from math import sqrt
print(f'RMSE: {sqrt(mean_squared_error(Y_test, Y_best_pred))}')
```
