```python
#Ahora vamos a retirar los datos atÃ­picos de las variables en cuestiÃ³n
df = df[(df['Kg malta'] >= 17370.25) & (df['Kg malta'] <= 18700)]
df = df[(df['Bultos Malta caramelo'] >= 6) & (df['Bultos Malta caramelo'] <= 26)]
df = df[(df['Color malta caramelo [EBC]'] >= 109) & (df['Color malta caramelo [EBC]'] <= 165)]
#df = df[(df['Color cocimiento'] >= 11.4) & (df['Color cocimiento'] <= 17)]
df = df[(df['Color ponderado silo'] >= 3.214) & (df['Color ponderado silo'] <= 4.8)]
```

```python
df.shape
```

```python
#Ahora voy a retirar las columnas de lote y silo   
df = df.drop(columns=['Lote malta caramelo', 'Silo','Kg malta'], axis=1)
```

```python
from utils import analisis_boxplot
analisis_boxplot(df,'Color cocimiento')
#Se cuenta con los datos sin atÃ­picos no deseados.
```

```python
#Ahora debo dividir los datos
from sklearn.model_selection import train_test_split
#AquÃ­ saco la columna que voy a predecir
X = df.drop(['Color cocimiento'], axis=1)
#AquÃ­ digo cuÃ¡l es la columna que voy a predecir
Y = df['Color cocimiento']
#AquÃ­ hago el split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train.head()
```

```python

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(max_depth=4, random_state=42)
model.fit(X_train,Y_train)
```

```python
Y_pred= model.predict(X_test)
Y_pred
```

```python
#ahora debo evaluar el modelo
from sklearn.metrics import mean_squared_error, r2_score

print(f'MSE: {mean_squared_error(Y_test, Y_pred)}')
print(f'R2: {r2_score(Y_test, Y_pred)}')

from math import sqrt
print(f'RMSE: {sqrt(mean_squared_error(Y_test, Y_pred))}')

#Ahora sacarÃ© las mismas metricas con los datos de entrenamiento
Y_train_pred = model.predict(X_train)
print(f'MSE train: {mean_squared_error(Y_train, Y_train_pred)}')
print(f'R2 train: {r2_score(Y_train, Y_train_pred)}') 
print(f'RMSE train: {sqrt(mean_squared_error(Y_train, Y_train_pred))}')
```

```python
import matplotlib.pyplot as plt


# Crear el grÃ¡fico
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--', lw=2)
plt.xlabel('Valores reales')
plt.ylabel('Predicciones')
plt.title('Predicciones vs Valores reales')
plt.show()
```

```python
importancia = pd.DataFrame({
    'caracteristica': X_train.columns,
    'importancia': model.feature_importances_
})
importancia = importancia.sort_values('importancia', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(importancia['caracteristica'], importancia['importancia'])
plt.xticks(rotation=45)
plt.xlabel('CaracterÃ­sticas')
plt.ylabel('Importancia')
plt.title('Importancia de caracterÃ­sticas')
plt.tight_layout()
plt.show()
```

```python
residuos = Y_test - Y_pred

plt.figure(figsize=(10, 6))
plt.scatter(Y_pred, residuos)
plt.axhline(y=0, color='r', linestyle='--')
plt.axhline(y=1, color='g', linestyle='--')  # LÃ­nea en y=2
plt.axhline(y=-1, color='g', linestyle='--') # LÃ­nea en y=2
plt.axhline(y=0.5, color='y', linestyle='--')  # LÃ­nea en y=2
plt.axhline(y=-0.5, color='y', linestyle='--') # LÃ­nea en y=2
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.title('GrÃ¡fico de residuos')
plt.show()
```

```python
#Ahora quiero ver el porcentaje de residuos que estÃ¡n por encima de 1 y por debajo de -1
porcentaje_residuos = ((residuos > 1).sum() + (residuos < -1).sum()) / len(residuos) * 100
print(f'Porcentaje de residuos por encima de 1 o por debajo de -1: {porcentaje_residuos:.2f}%')
```
